```python
# analysis for Outliers
# For the columns having we will remove them and consider only those values which lie between 25th precentile and 75 percentile.
sns.boxplot(data['clip_views'])
plt.title('outliers analysis')
plt.show()
```

```python
# Removing the outliers
Q1 = np.percentile(data['clip_views'], 25, interpolation = 'midpoint')  
Q2 = np.percentile(data['clip_views'], 50, interpolation = 'midpoint')  
Q3 = np.percentile(data['clip_views'], 75, interpolation = 'midpoint')     
IQR = Q3 - Q1 
low_lim = Q1 - 1.5 * IQR 
up_lim = Q3 + 1.5 * IQR
print('low limit:', low_lim) 
print('upper limit:', up_lim)

data=data[data.clip_views<up_lim]
data=data[data.clip_views>low_lim]
```
